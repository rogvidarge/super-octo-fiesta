```python
# The factorial of a number N is defined as the product of all values from 1 to N (included) and is written as N!.
# So 3! is the same as 1 x 2 x 3 = 6.

def factorial(n):
    result = 1
    for i in range(2, n):
        result *= i
    return result

for i in [0, 5, 6]:
    print(factorial(i))

# Expected output: 
# 1
# 120
# 720
```
